## Preparing the data for the ML models

````python
ohe = OneHotEncoder(sparse_output=False) # Turning off sparce_output to maintaining this like a dataframe
ode = OrdinalEncoder
SI  = SimpleImputer(strategy='most_frequent')
````

### Preparing the columns that we gonna apply ohe
- ODE is to deal with ranking data, like the family_Size_grouped, that involves different group sizes
- OHE is to data that don't involves ranking

````python
train_df.keys()
````

````python
ode_cols = ['Family_Size_Grouped'] # Data with ranking
ohe_cols = ['Sex', 'Embarked'] # Data without ranking
````

````python
X = train_df.drop(['Survived'], axis=1)
y = train_df['Survived']
X_test = test_df.drop(['Age_Cut', 'Fare_Cut'], axis=1)
````

## Building the pipelines

### Ordinal Pipeline
- SimpleImputer(strategy='most_frequent')
    - Look for missing values in categorical columns with the most frequent value (mode) of the respective column.
    - Useful strategy for columns with discrete data that repeats frequently
- OrdinalEncoder(...)
    - Converts categories to integers (e.g 'low', 'medium', 'heigh' -> 0, 1, 2)
    - The parameter:
        - handle_unknown='use_encoded_value' instructs the encoder not to throw an error if a new category appears during inference.
        - unknown_value=-1 specifies that unseen categories will be mapped to -1.

````python
ordinal_pipeline = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('ord', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
])
````

### OHE Pipeline
- SimpleImputer(strategy='most_frequent')
    - Same as above
- OneHotEncoder(...)
    - Converst each category to a binary column (0 or 1)
    - For example: ['red', 'green', 'blue'] -> thhree columns: [1, 0], [0, 1, 0], [0, 0, 1]
    - Parameters:
        - handle_unknown='ignore': New categories during inference will be ignored (all columns result in 0).
        - sparse_output=False: returns a dense_array (useful for debugging or integration with pipeplines that do not accept sparse matrices)

````python
ohe_pipeline = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('one-hot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])
````

## Building Transformers

````python
train_df.keys()
````

### What is it 'ColumnTransfomer'?
- Allows to define a set of specific transformations for groups of columns. Internally, it builds a pipeline: Each subset of parallel columns is processed with the indicated transformer, and the results are concatenated at the end.
### Why used ?
- Without ColumnTransformer, you would have to manually process each column type and then merge or concat the results -- something that is tedious, error-prone, and difficult to encapsulate in a complete pipeline. 
- With it:
    - The preprocessing logic is clear and centralized
    - It can be integrated with pipelines (scikit-learn Pipeline)
    - It allows direct use with fit, transform, fit_transform and GridSearchCV.

#### General structure:
```python
ColumnTransformer(transformers=[
    ('block_name', transformer, columns_list),
    ...
    ], remainder='drop' or 'passthrough')
```
- block_name:
    - Internal identifier name of the transformer
- Transformer:
    - Any object that has .fit() and .transform() (e.g SimpleImputer, StandardScaler, Pipeline, etc.)
- Column_list:
    - Columns from the dataset to be transformed with that block.
- Remainder:
    - 'drop':
        - Columns not reserved will be discarded
    - 'passthrough':
        - Columns not specifically processed will be copied to the final outupt
- n_jobs:
    - Paralelization. -1=Use all the ncleus

````python
col_trans = ColumnTransformer(transformers=[
    ('impute', SI, ['Age']),
    ('ord_pipeline', ordinal_pipeline, ode_cols),
    ('ohe_pipeline', ohe_pipeline, ohe_cols),
    ('passthrough', 'passthrough', ['Pclass', 'TicketNumberCounts', 'Cabin_Assigned', 'Name_Size', 'Age', 'Fare'])
],remainder='drop', n_jobs=-1)
````
